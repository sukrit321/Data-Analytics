# <font color=blue> Naive Bayes 

### <font color=byellow> ใช้ไฟล์ ContactLens.csv ไฟล์ ประกอบด้วยคุณลักษณะดังนี้
- ### <font color=blue> Age (กลุ่มอายุคนไข้) : young, pre-presbyopic, presbyopic (สายตาผู้สูงอายุ)
- ### <font color=blue> Prescript (ลักษณะการมองเห็น): myope(สายตาสั้น), hypermetrope(สายตายาว)
- ### <font color=blue> Astigmatism (สายตาเอียง): no, yes
- ### <font color=blue> TearProdRate (น้ำตา) : reduced, normal
- ### <font color=blue> ContactLens มี 3 ประเภท คือ none, soft, hard

### <font color=bgreen> ลุงตู่มาตรวจสายตาเพื่อจะได้ทราบว่าต้องใส่ contact lens หรือไม่รายละเอียดมีดังนี้
- ### <font color=blue> ลุงตู่ => young, myope, yes, reduced

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("ContactLens.csv")
```

```python
#เขียนโปรแกรมโดยใช้วิธี naive bayes เพื่อช่วยให้คำแนะนำเบื้องต้นว่าลงตู่น่าจะใส่ contact lens ประเภทไหน 
print(data)
```

```python
n_none = len(data[data['ContactLens'] == "none"])

n_soft = len(data[data['ContactLens'] == "soft"])

n_hard = len(data[data['ContactLens'] == "hard"])

total_ContactLens = len(data['ContactLens'])

print(n_none, n_soft, n_hard, total_ContactLens)
```

```python
n_young_none = len(data[(data['ContactLens'] == "none") & (data['Age'] ==  "young")])
n_pre_presbyopic_none = len(data[(data['ContactLens'] == "none") & (data['Age'] ==  "pre-presbyopic")])
n_presbyopic_none = len(data[(data['ContactLens'] == "none") & (data['Age'] ==  "presbyopic")])

n_myope_none = len(data[(data['ContactLens'] == "none") & (data['Prescript'] ==  "myope")])
n_hypermetrope_none = len(data[(data['ContactLens'] == "none") & (data['Prescript'] ==  "hypermetrope")])

n_yes_none = len(data[(data['ContactLens'] == "none") & (data['Astigmatism'] ==  "yes")])
n_no_none = len(data[(data['ContactLens'] == "none") & (data['Astigmatism'] ==  "no")])

n_reduced_none = len(data[(data['ContactLens'] == "none") & (data['TearProdRate'] ==  "reduced")])
n_normal_none = len(data[(data['ContactLens'] == "none") & (data['TearProdRate'] ==  "normal")])
```

```python
n_young_soft = len(data[(data['ContactLens'] == "soft") & (data['Age'] ==  "young")])
n_pre_presbyopic_soft = len(data[(data['ContactLens'] == "soft") & (data['Age'] ==  "pre-presbyopic")])
n_presbyopic_soft = len(data[(data['ContactLens'] == "soft") & (data['Age'] ==  "presbyopic")])

n_myope_soft = len(data[(data['ContactLens'] == "soft") & (data['Prescript'] ==  "myope")])
n_hypermetrope_soft = len(data[(data['ContactLens'] == "soft") & (data['Prescript'] ==  "hypermetrope")])

n_yes_soft = len(data[(data['ContactLens'] == "soft") & (data['Astigmatism'] ==  "yes")])
n_no_soft = len(data[(data['ContactLens'] == "soft") & (data['Astigmatism'] ==  "no")])

n_reduced_soft = len(data[(data['ContactLens'] == "soft") & (data['TearProdRate'] ==  "reduced")])
n_normal_soft = len(data[(data['ContactLens'] == "soft") & (data['TearProdRate'] ==  "normal")])
```

```python
n_young_hard = len(data[(data['ContactLens'] == "hard") & (data['Age'] ==  "young")])
n_pre_presbyopic_hard = len(data[(data['ContactLens'] == "hard") & (data['Age'] ==  "pre-presbyopic")])
n_presbyopic_hard = len(data[(data['ContactLens'] == "hard") & (data['Age'] ==  "presbyopic")])

n_myope_hard = len(data[(data['ContactLens'] == "hard") & (data['Prescript'] ==  "myope")])
n_hypermetrope_hard = len(data[(data['ContactLens'] == "hard") & (data['Prescript'] ==  "hypermetrope")])

n_yes_hard = len(data[(data['ContactLens'] == "hard") & (data['Astigmatism'] ==  "yes")])
n_no_hard = len(data[(data['ContactLens'] == "hard") & (data['Astigmatism'] ==  "no")])

n_reduced_hard = len(data[(data['ContactLens'] == "hard") & (data['TearProdRate'] ==  "reduced")])
n_normal_hard = len(data[(data['ContactLens'] == "hard") & (data['TearProdRate'] ==  "normal")])
```

```python
p_none = n_none/total_ContactLens
p_soft = n_soft/total_ContactLens
p_hard = n_hard/total_ContactLens

p_young_none = n_young_none/n_none
p_pre_presbyopic_none = n_pre_presbyopic_none/n_none
p_presbyopic_none = n_presbyopic_none/n_none

p_myope_none = n_myope_none / n_none
p_hypermetrope_none = n_hypermetrope_none / n_none

p_yes_none = n_yes_none / n_none
p_no_none = n_no_none / n_none

p_reduced_none = n_reduced_none / n_none
p_normal_none = n_normal_none / n_none
```

```python
p_young_soft = n_young_soft/n_soft
p_pre_presbyopic_soft = n_pre_presbyopic_soft/n_soft
p_presbyopic_soft = n_presbyopic_soft/n_soft

p_myope_soft = n_myope_soft / n_soft
p_hypermetrope_soft = n_hypermetrope_soft / n_soft

p_yes_soft = n_yes_soft / n_soft
p_no_soft = n_no_soft / n_soft

p_reduced_soft = n_reduced_soft / n_soft
p_normal_soft = n_normal_soft / n_soft
```

```python
p_young_hard = n_young_hard/n_hard
p_pre_presbyopic_hard = n_pre_presbyopic_hard/n_hard
p_presbyopic_hard = n_presbyopic_hard/n_hard

p_myope_hard = n_myope_hard / n_hard
p_hypermetrope_hard = n_hypermetrope_hard / n_hard

p_yes_hard = n_yes_hard / n_hard
p_no_hard = n_no_hard / n_hard

p_reduced_hard = n_reduced_hard / n_hard
p_normal_hardt = n_normal_hard / n_hard
```

```python
p_young_myope_yes_reduced_none = p_young_none * p_myope_none * p_yes_none * p_reduced_none * p_none

p_young_myope_yes_reduced_soft = p_young_soft * p_myope_soft * p_yes_soft * p_reduced_soft * p_soft

p_young_myope_yes_reduced_hard = p_young_hard * p_myope_hard * p_yes_hard * p_reduced_hard * p_hard

print(p_young_none, p_myope_none, p_yes_none, p_reduced_none, p_none)
print(p_young_soft, p_myope_soft, p_yes_soft, p_reduced_soft, p_soft)
print(p_young_hard, p_myope_hard, p_yes_hard, p_reduced_hard, p_hard)

print("p_young_myope_yes_reduced_none", p_young_myope_yes_reduced_none)
print("p_young_myope_yes_reduced_soft", p_young_myope_yes_reduced_soft)
print("p_young_myope_yes_reduced_hard", p_young_myope_yes_reduced_hard)
```

### <font color=green> ตอบช่องนี้ว่าลุงตู่น่าจะใส้ contact lens แบบไหน ให้เหตุผลอธิบาย ใส่แบบ none เพราะ ค่า P(ContactLens = none|new) (=0.0183) มีค่ามากกว่า P(ContactLens = soft|new) (=0.000) และ P(ContactLens = hard|new)ดังนั้นโมเดลจึงทำนายว่าข้อมูลตัวอย่างนี้มีค่าคลาส ContactLens = none
