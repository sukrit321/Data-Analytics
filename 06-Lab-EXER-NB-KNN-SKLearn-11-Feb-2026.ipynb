```python
import pandas as pd
import numpy as np
#Import Library of Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
#Import Library of K-Nearest Neighbor model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```python
#1. ไฟล์ข้อมูล .csv ที่ให้มา ให้นักศึกษาใช้ Library of Gaussian Naive Bayes model
#เพื่อ predict ข้อมูลใหม่ที่เป็น rainy, hot, high, weak ได้คำตอบ class เป็นอะไร
play_tennis = pd.read_csv("play_tennis.csv")
print(play_tennis)

le = preprocessing.LabelEncoder()
# Converting string labels into numbers.
play_tennis['outlook'] =  le.fit_transform(play_tennis['outlook'])
play_tennis['temperature'] = le.fit_transform(play_tennis['temperature'])
play_tennis['humidity'] = le.fit_transform(play_tennis['humidity'])
play_tennis['windy'] = le.fit_transform(play_tennis['windy'])



#print(play_tennis[['outlook', 'temperature' ,'humidity'  , 'windy', 'play']])
#outlook = list(zip(play_tennis['outlook'],outlook_encoded))
#temperature = list(zip(play_tennis['temperature'], temperature_encoded))
#humidity = list(zip(play_tennis['humidity'], humidity_encoded ))
#windy = list(zip(play_tennis['windy'], windy_encoded))
print(play_tennis)

```

```python
data = pd.DataFrame()

data['outlook'] = [1]
data['temperature'] = [1]
data['humidity'] = [0]
data['windy'] = [1]

model = GaussianNB()
model.fit(play_tennis[['outlook', 'temperature', 'humidity', 'windy']], play_tennis['play'])
predicted= model.predict(data) 
print("Predicted Value:", predicted)
```

```python
#2.ไฟล์ข้อมูล .csv ที่ให้มา ให้นักศึกษาใช้ KNN model เมื่อ K=3
#เพื่อ predict ข้อมูลใหม่ที่เป็น rainy, hot, high, weak ได้คำตอบ class เป็นอะไร

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(play_tennis[['outlook', 'temperature', 'humidity', 'windy']], play_tennis['play'])

predicted= knn.predict(data) 
print("Predicted Value:", predicted)
```

```python
#3. มีข้อมูลใหม่ 2 ชุดคือ 
#outlook -> rainy, sunny
#temperature -> cool, mild
#humidity -> high, normal
# windy -> weak, strong
#ให้นักศึกษาใช้ NB เพื่อ predict ข้อมูลใหม่ว่าอยู่ใน class อะไร

data_1 = pd.DataFrame()
data_1['outlook'] = [1,2]
data_1['temperature'] = [1,2]
data_1['humidity'] = [0,1]
data_1['windy'] = [1,0]

model = GaussianNB()
model.fit(play_tennis[['outlook', 'temperature', 'humidity', 'windy']], play_tennis['play'])
predicted= model.predict(data_1) 
print("Predicted Value:", predicted)
```

```python
#4. มีข้อมูลใหม่ 2 ชุดคือ 
#outlook -> rainy, sunny
#temperature -> cool, mild
#humidity -> high, normal
# windy -> weak, strong
#ให้นักศึกษาใช้ KNN (เมื่อ K=3) เพื่อ predict ข้อมูลใหม่ว่าอยู่ใน class อะไร

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(play_tennis[['outlook', 'temperature', 'humidity', 'windy']], play_tennis['play'])

predicted= knn.predict(data_1) 
print("Predicted Value:", predicted)
```
