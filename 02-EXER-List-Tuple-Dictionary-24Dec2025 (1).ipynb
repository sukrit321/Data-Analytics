```python
#14. สร้างตัวแปร set ชุดที่่1 ที่มีสมาชิก {10, 20, 30, 40, 50} และตัวแปร set ชุดที่ 2 ที่มีสมาชิก {30, 40, 50, 60, 70}
#เชียน python code เพื่อนำข้อมูลสมาชิกของ set ชุดที่ 1 และชุดที่ 2 
#มาใช้เพื่อให้สมาชิกใน set ที่ประกอบไปด้วย {70, 40, 10, 50, 20, 60, 30}
# Hints : ข้อนี้ให้ใช้ union buit-in function
```

```python
a = {10, 20, 30, 40, 50}
b = {30, 40, 50, 60, 70}
print(a.union(b))
```

```python
#15. จากคำตอบข้อ 14. เรียงข้อมูลจากมากไปน้อย  และแสดง data type ของข้อมูลที่เรียงจากมากไปน้อยดังกล่าว
```

```python
c = a.union(b)
c = sorted(c, reverse=True)
print(c)
```
