### Missing Values

```python
flights = pd.read_csv("flights.csv")
flights.tail()
```

```python
# เลือกข้อมูลที่มี missing value อย่างน้อย 1 ตัวในแต่ละแถว
flights[flights.isnull().any(axis=1)].head()
```

```python
# เลือก arr_delay ที่มีค่า missing value
flights1 = flights[ flights['arr_delay'].isnull( )]
flights1.head()
```

```python
# เลือกข้อมูลทั้งหมดที่ตัด missing value ออกไป
flights2_nomissing = flights.dropna()
flights2_nomissing
```

```python
#df2=df.dropna(subset=['Courses','Duration'])
drop_missing_dep_delay = flights.dropna(subset=['dep_delay'])
drop_missing_dep_delay
```

```python
# แสดงค่า missing value
#print("any missing", flights[ flights['dep_delay'].isnull( )].any())
print("any missing", flights[ flights['dep_delay'].isnull( )])
```

```python
# เปลี่ยนค่า missing values ของ dep_delay ด้วยค่าศูนย์
print("any missing after fill missing",flights['dep_delay'].fillna(0).isnull().any())
#nomiss =flights['dep_delay'].fillna(0)
#nomiss.isnull().any()  # ตรวจสอบว่า nomiss เป็น null หรือไม่ จะให้ค่า True or False ถ้าเป็น null จะมีค่าเป็น true
```

```python
# เปลี่ยนค่า missing values ของ dep_delay ด้วยค่าเฉลี่ยของ dep_delay
mean_dep_delay = flights['dep_delay'].mean() #หรือจะใช้คำสั่ง mean_dep_delay = np.mean(flights['dep_delay'])
average_nomiss =flights['dep_delay'].fillna(mean_dep_delay) #หรือจะใช้คำสั่ง flights['dep_delay'].fillna(flights['dep_delay'].mean())
average_nomiss
```

```python
# นับจำนวน missing values ที่อยู่ในแต่ละ dep_delay, arr_delay attribute
flights[['dep_delay','arr_delay']].isnull().sum()
```

```python
#นับจำนวน missing value ของ dep_delay และ arr_delay ทั้งหมด
flights[['dep_delay','arr_delay']].isnull().sum().sum() 
```

```python
# Find the number of non-missing values in each column
# นับจำนวนที่ไม่ใช่ missing value ของแต่ละ attribute
flights.count()
```
