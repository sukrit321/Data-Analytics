```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. อ่านไฟล์ข้อมูลชื่อ flight.csv 
flights = pd.read_csv("flights.csv")
```

```python
#2 แสดงผลสถิติเชิงพรรณา
flights.describe()
```

```python
#3. แสดงผล attribute name ของ distance hour minute ตั้งแต่ แถวที่ 50 ไปอีก 20 แถว
flights[["distance","hour","minute"]][50:70]
```

```python
#4. แสดงผลเฉพาะ attribute name ชื่อ flight
flights[["flight"]]
```

```python
#5.แสดงผลสถิติแยกตามชื่อสนามบินปลายทาง (dest)
print(flights.groupby(['dest']).ngroups)
flights.groupby(['dest']).describe()
```

```python
#6. แสดงผลค่าเฉลี่ยของระยะทางที่บิน (distance)
flights["distance"].mean()
```

```python
#7. ให้แสดงผลค่าเฉลี่ยของระยะทางที่บิน (distance) แยกตามกลุ่มเที่ยวบิน (flight)
#print(flights.groupby(['flight','distance']).ngroups)
flights.groupby("flight")["distance"].mean()
```

```python
#8. ให้พิมพ์ค่าเฉลี่ยของการออกไม่ตรงเวลาของสายการบิน (dep_delay)
flights["dep_delay"].mean()
```

```python
#9. ต้องการแสดงผลเฉพาะสนามบินปลายทางที่ชื่อ DEN
flights[flights["dest"] == "DEN"]
```

```python
#10. ต้องการแสดงผลเฉพาะ attribute dep_delay และ arr_delay ของสนามบินปลายทางที่ชื่อ DEN
flights[flights["dest"] == "DEN"][["dep_delay","arr_delay"]]
```

```python
#11. ตรวจสอบว่าข้อมูลที่อยู่ในไฟล์ csv มี missing value หรือไม่
flights[ flights['arr_delay'].isnull( )]
```

```python
#12. นับจำนวน missing value ของ dep_time 
flights[['dep_time']].isnull().sum()
```

```python
#13. จากข้อ 12 เลือกวิธีการที่เหมาะสมสำหรับจัดการกับค่า missing value 
mean_dep_delay = flights['dep_time'].mean()
average_nomiss =flights['dep_time'].fillna(mean_dep_delay)
average_nomiss
```

```python
#14.เลือกวิธีการที่เหมาะสมสำหรับจัดการกับค่า missing value ของ tailnum และแสดงให้เห็นว่าไม่มีค่า missing ของ attribute นี้
max_tail_tailnum = flights['tailnum'].value_counts().idxmax()
max_tail_tailnum = flights['tailnum'].fillna('N328AA').isnull().any()
max_tail_tailnum
```

```python
#15.เลือกวิธีการที่เหมาะสมสำหรับจัดการกับค่า missing value ของ tailnum และแสดงให้เห็นว่าไม่มีค่า missing ของ attribute นี้ ใช้ mode
mode_tail_tailnum = flights['tailnum'].mode()
mode_tail_tailnum = flights['tailnum'].fillna('N328AA').isnull().any()
mode_tail_tailnum 
```
